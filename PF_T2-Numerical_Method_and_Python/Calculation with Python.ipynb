### FDM: An example first order ordinary differential equation

(The following part is from one of the resources given before this tutorial)

Suppose we have a first order ODE:

$$\frac{\mathrm{d}\,u(t)}{\mathrm{d}\,t}=\lambda u(t)+f(t),$$

with the initial value 

$$u(0)=u_0,$$

here $f(t)$ is a known function, and $u(t)$ is the function we are going to solve. This is a heat transfer model with f(t), the time dependent heat source.

Let's get started, but from the most basic one: numerical differential. 


#### Numerical Differential

Take a look at the following identity:

$$f(x) = \sin(x),\space f'(x) = \cos(x),$$

But this is an analytical solution, not a numerical one. How to differentiate the $f(x)$ numerically? It should be quite easy, let's see its numerical solution with forward, backward and central differencing.

As a (maybe) revision, here is how these methods work. With forward differencing, we mean taking 
$$\frac{\mathrm{d}\,f}{\mathrm{d}\,x} = \frac{f_(x+h) - f(x) }{h};$$
So for backward one, we got
$$\frac{\mathrm{d}\,f}{\mathrm{d}\,x} = \frac{f_(x) - f(x-h) }{h};$$
And the central differencing is:
$$\frac{\mathrm{d}\,f}{\mathrm{d}\,x} = \frac{f_(x+h/2) - f(x-h/2) }{h}.$$
I have to say it's quite straightforward, with $h$ takes responsibility of $\mathrm{d}/,x$. We can foresee that if $h$ is small, the result should be a good approximation.
But how good it can be? Now let's implement them to test it.

````python
## We have imported numpy and matplotlib, so we don't need them here.

def ForwardSin(x,h): #ForwardSin is a function of x and h
    return (np.sin(x+h)-np.sin(x))/h

x=np.linspace(0,2*np.pi,501) #define a set of input values
#define where plots will appear
plt.plot(x,np.cos(x),'-b', label="$\\cos(x)$") #plot f'(x)=cos(x)
plt.plot(x,ForwardSin(x,0.01),'--r', label="$(\\sin(x+h)-\\sin(x))/h$, $h=0.01$") #plot f'(x)=(f(x+h)-f(x))/h
plt.xlabel('x') #label horizontal axis
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.); #plot legend
````

Then let's see what will happen if we change $h$ to 1:

````python
plt.plot(x,np.cos(x),'-b', label="$\\cos(x)$") #plot f'(x)=cos(x)
plt.plot(x,ForwardSin(x,1),'--r', label="$(\\sin(x+h)-\\sin(x))/h$, $h=1$") #plot f'(x)=(f(x+h)-f(x))/h
plt.xlabel('x') #label horizontal axis
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.); #plot legend
````

Then, let's try backward differencing and central differencing, with different $h$ value:

````python
#---------------------------#
### Backward Differencing ###
#---------------------------#
def BackwardSin(x,h): #ForwardSin is a function of x and h
    return (np.sin(x)-np.sin(x-h))/h
plt.plot(x,np.cos(x),'-b', label="$\\cos(x)$") #plot f'(x)=cos(x)
plt.plot(x,BackwardSin(x,0.01),'--r', label="$(\\sin(x)-\\sin(x-h))/h$, $h=0.01$") #plot f'(x)=(f(x)-f(x-h))/h
plt.plot(x,BackwardSin(x,1),'--g', label="$(\\sin(x)-\\sin(x-h))/h$, $h=1$") #plot f'(x)=(f(x)-f(x-h))/h
plt.xlabel('x') #label horizontal axis
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.); #plot legend
````

````python
#---------------------------#
### Central Differencing ###
#---------------------------#
def CentralSin(x,h): #CentralSin is a function of x and h
    return (np.sin(x+h/2)-np.sin(x-h/2))/h
plt.plot(x,np.cos(x),'-b', label="$\\cos(x)$") #plot f'(x)=cos(x)
plt.plot(x,CentralSin(x,0.01),'--r', label="$(\\sin(x+h/2)-\\sin(x-h/2))/h$, $h=0.01$") #plot f'(x)=(f(x+h/2)-f(x-h/2))/h
plt.plot(x,CentralSin(x,1),'--g', label="$(\\sin(x+h/2)-\\sin(x-h/2))/h$, $h=1$") #plot f'(x)=(f(x+h/2)-f(x-h/2))/h
plt.xlabel('x') #label horizontal axis
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.); #plot legend
````

You may noticed that these methods will cause different behaviour when $h$ is too big, while they all perform well when we have a fine $h$.

#### Euler Method for Differential Equation

The equation we have is:
$$\frac{\mathrm{d}\,u(t)}{\mathrm{d}\,t}=\lambda u(t)+f(t).$$
By forward Euler method, we have:
$$\frac{u(t_i) - u(t_{i-1})}{h}=\lambda u(t_{i-1})+f(t_{i-1}),$$
Let's denote $u(t_i)$ as $u_i$ and $f(t_{i-1})$ ad $f_{i-1}$, then if we have got $u_{i-1}$  and $f(t_{i-1})$ already (that is, we get the results in time step $t_{i-1}$), we can solve $u_i$ as follows, by arranging the equation above:
$$u_i = (h\lambda+1)u_{i-1} + h f_{i-1}$$
And, now we get the initial value of $u(0) = u_0$, and the function $f$ is known, then we can solve this differential step by step.

Ah, by the way, if you are trying to use the backforward Euler method, that is:
$$\frac{u(t_i) - u(t_{i-1})}{h}=\lambda u(t_{i})+f(t_{i}).$$
(Please notice that here we use $f(t_{i})$, but usually, we just can't get the value of $f(t)$ in the next timestep as it may coupled with u. However, in that case, we still can use $f(t_{i-1})$ to approach the result as a work around.)

From this equation, with the notation we made before, you might get the iteration equation as follows (which is quite different from the explicit (forward) method):
$$u_i = \frac{u_{i-1} + h f_{i}}{1-\lambda h}.$$
But let's stick on the forward Euler method for now.

Suppose the function $f(t) = \sin(t)$, $\lambda = 0.1$, initial value is $u_0=1$, and we choose time step $h$ as 0.01.

We will plot this result with the numerical method.

````python
import numpy as np
from matplotlib import pyplot as plt
u=[1] #set initial value for u. 
#u is a list of values that initially has length 1, but which will get longer at each step in the for loop
t0=0 #set initial time point
tN=10 #set final time point
N=1000 #set number of sections to divide time domain into
h=(tN-t0)/N #calculate the width of each section

t=np.linspace(t0,tN,N+1) #set up an array of input time values
T=np.linspace(t0+h,tN,N) #set up an array of time values excluding t=0

for tj in T:
    u.append(h*0.1*u[-1] + h*np.sin(tj) + u[-1]) #for each time step, calculate the value of u. Notice the u[-1] means the last number of u.
    
plt.plot(t,(201*np.exp(0.1*t)-100*np.cos(t)-10*np.sin(t))/101,'-b', label="Exact solution") #plot exact solution
plt.plot(t,u,'--r',label="Numerical solution") #plot u
plt.xlabel('t') #label horizontal axis
plt.ylabel('u(t)'); #label vertical axis
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.); #plot legend
````

Great, now you should have your way to solve a first order ODE. What about second order? The answer is: *Solve It Twice*.

Suppose we have the following second order ODE:
$$\frac{\mathrm{d}^2\,u(x)}{\mathrm{d}\,x^2}=f(x),$$
with known $f(x)$, $u(0)$ and $(u'(0))$.

we can split this equation as:
$$ \frac{\mathrm{d}}{\mathrm{d}\,x} \frac{\mathrm{d} u(x)}{\mathrm{d}\,x} = f(x).$$
Now maybe you already know that. Just solve the left part first, taking $\frac{\mathrm{d} u(x)}{\mathrm{d}\,x}$ as a single function: $v = u' = \frac{\mathrm{d} u(x)}{\mathrm{d}\,x}$, then you get:
$$ \frac{\mathrm{d}\,v}{\mathrm{d}\,x} = f(x),$$
and 
$$ \frac{\mathrm{d}\,u}{\mathrm{d}\,x} = v(x).$$
Now, $v(0) = u'(0)$, and you know what to do.

Let's see an example, with $u(0) = 1$, $u'(0) = v(0) = 2$ and $f(x) = 6x$. The exact solution is:

$$u(x) = x^3 + 2x+1$$

### Integrate in 2D

Now we are going to integrate some function in 2D space. But how to define a two dimensional space? Let't try to implement it with `class`.

- Functions' value

So, there are at least two way to implement this:
1. Create a 2D array, the array elements contain all data in one point, and positional data are binded with the subscript of the array.
2. Create multiple 2D arrays, each array contains one kind of data, and positional data are binded with the subscript of the array.

We will try the first method. Who interested in the second method are welcomed to try themselves.

So, let't create a class, name it as `Mesh`.

````python
class Mesh:
    def __init__(self, x_length, dx, y_length=None):
        self.x_length = x_length  # The length in x direction
        # If y_length is not given, just assume it is the same as x_length.
        if y_length is None:
            self.y_length = x_length
        # Else, use the given value.
        else:
            self.y_length = y_length
        # Initialize the mesh points
        self.dx = dx  # The x direction length step, we with the grid is divided into square, so there is no dy.
        # Define the size of the mesh 
        ## Notice that size is different with length here, with "length" a float number indicates edge's length
        ## While "size" here indicates the number of points we take in the edges, which should be a int number.
        self.N_x = int(self.x_length / dx)
        self.N_y = int(self.y_length / dx)
        grid = []
        one_row = np.zeros(self.N_x)
        for i in range(self.N_y):
            grid.append(one_row)
        self.nodes = np.array(grid)
        # The self.nodes is now a x_length*y_length matrix, filling with zero.
        # others #

# So we have a small square mesh, with length 1 and x step 0.01. The grids are filled with 0 as default value.
mesh_a = Mesh(1, 0.01) 
print(f"The X dimension of mesh_a contains {mesh_a.nodes[0].size} points")
print(f"The Y dimension of mesh_a contains {mesh_a.nodes[:,0].size} points")
mesh_b = Mesh(0.255,0.1,y_length=2)
print(f"The X dimension of mesh_b contains {mesh_b.nodes[0].size} points")
print(f"The Y dimension of mesh_b contains {mesh_b.nodes[:,0].size} points")
````

Now let's define a function to be integrated, say, $$z = (x+y)^2$$ on our a $[0,1]\times[0,1]$ domain.

Its integral in domain $[0,1]\times[0,1]$ is $7/6 \approx 1.166667$. 

````python
def assign_val(self, func): 
    for i in range(self.N_x):
        for j in range(self.N_y):
            x = i*self.dx
            y = j*self.dx
            self.nodes[i][j] = func(x,y)
## Add this method to Mesh class.
Mesh.assign_val = assign_val
## Here let's visualize the result:
def square_x_y (x,y):
    return (x+y)**2

mesh_plot = Mesh(1, 0.01)
mesh_plot.assign_val(square_x_y)

from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib.cm import coolwarm
x_axis = np.linspace(0,1,mesh_plot.N_x)
y_axis = np.linspace(0,1,mesh_plot.N_y)
X,Y = np.meshgrid(x_axis, y_axis)
Z = mesh_plot.nodes

fig = plt.figure(figsize=(14,6))

# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 1, 1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
````

Ok, Now let's integrate it. Things are, actually really easy. All you need to do is sum the function value in each node, and multiply the result with the grid size, that is, `dx * dx`:

````python
def RiemannIntegral(mesh:Mesh)->float:
    sum = 0
    for row in mesh.nodes:
        for val in row:
            sum = sum + val
    return sum*mesh.dx*mesh.dx

mesh_calc = Mesh(1,0.001)
mesh_calc.assign_val(square_x_y)
integral_1 = RiemannIntegral(mesh_calc)
print(integral_1)
````

Alright, the result seems to be better, but this time cost obviously more time. Is there any better way to do integral?

The answer is: Yes.

**Notice: The following part is just for performing the algorithm. In practice, you just use the simplest way.**

We have quadrature rules:

$$ \int_a^b f(x) \mathrm{d}\, x \approx \frac{h}{2} \left\{ f(a) +  2\sum_{i=0}^{n-1}f(x_{i}) + f(b) \right\} $$

But this is for single variable function integral. However, we can calculate one dimension first, for example, x dimension, to
form a list, then take the new list to be another function to integrate. Or say, you integrate twice. Here's how you can 
do that.


````python
def QuadratureIntegral(mesh: Mesh) -> float:
    sum = 0
    row_list = []
    for row in mesh.nodes:
        row_sum = 0
        for val in row:
            row_sum +=  val
        row_sum -= (row[0]+row[-1])/2
        sum += row_sum
        row_list.append(row_sum)
    sum -=(row_list[0]+row_list[-1])/2
    return sum*mesh.dx*mesh.dx
integral_3 = QuadratureIntegral(mesh_calc)
print(integral_3)
````

Oh no the result is even worse. But that might be a template to perfom more complex algorithm.

Introducing: *Simpson Formula*
$$ \int_a^b f(x) \mathrm{d}\, x \approx \frac{h}{6} \left\{ f(a) + 4 \sum_{i=0}^{n-1}f(x_{i+1/2}) + 2\sum_{i=0}^{n-1}f(x_{i}) + f(b) \right\} $$

````python
def SimpsonIntegral(mesh: Mesh) -> float:
    sum = 0
    first_row_sum = 0
    last_row_sum = 0
    for i in range(mesh.N_x):
        row_sum = 0
        for j in range(mesh.N_y):
            if j%2==0:
                row_sum += mesh.nodes[i][j]*2
            else:
                row_sum += mesh.nodes[i][j]*4
        row_sum -= (mesh.nodes[i][0]+mesh.nodes[i][-1])
        
        if i ==0:
            first_row_sum = row_sum
        elif i == mesh.N_x -1:
            last_row_sum = row_sum
        
        if i % 2==0:
            sum += row_sum*2
        else:
            sum += row_sum*4
    sum -= (first_row_sum +last_row_sum)
    sum *= (mesh.dx*mesh.dx*4)/36
    return sum
integral_4 = SimpsonIntegral(mesh_calc)
print(integral_4)
````

It's better, right? Actually we have *Newton-Cotes* formula:
$$\int_a^b f(x) \mathrm{d}\, x \approx \frac{h}{90} \left\{ 7f(a) + 32 \sum_{i=0}^{n-1} f(x_{i+1/4}) + 12 \sum_{i=0}^{n-1} f(x_{i+1/2}) + 32\sum_{i=0}^{n-1} f(x_{i+3/4}) + 14\sum_{i=0}^{n-1} f(x_{i}) + 7f(b) \right\}$$

````python
def N_C_Integral(mesh: Mesh) -> float:
    sum = 0
    first_row_sum = 0
    last_row_sum = 0
    for i in range(mesh.N_x):
        row_sum = 0
        for j in range(mesh.N_y):
            if j % 4 == 0:
                row_sum += mesh.nodes[i][j] * 14
            elif j % 4 == 1 or j % 4 == 3:
                row_sum += mesh.nodes[i][j] * 32
            elif j % 4 == 2:
                row_sum += mesh.nodes[i][j] * 12
        row_sum -= (mesh.nodes[i][0]+mesh.nodes[i][-1])*7
        if i ==0:
            first_row_sum = row_sum
        elif i == mesh.N_x -1:
            last_row_sum = row_sum
        if j % 4 == 0:
            sum += row_sum * 14
        elif j % 4 == 1 or j % 4 == 3:
            sum += row_sum * 32
        elif j % 4 == 2:
            sum += row_sum * 12
    sum -= (first_row_sum +last_row_sum)
    sum *= (4*mesh.dx*4*mesh.dx)/8100
    return sum
integral_5 = N_C_Integral(mesh_calc)
print(integral_5)
````

These method achieves better and better result withoud changing the step length. You can futher optimize calculation speed of these calculation method by performing less multiplication.

### $\nabla$ and $\nabla^2$ 

Yes, you reached the final part of this notebook. We finally are facing the gradient and laplacian. With OOP, the class we created and FDM (if you forget, finite difference method), it's actually a easy thing to do.

Let's add a new class `Node` and re-shape our class `Mesh` here. Don't worry, the new definition will overwrite the old definition.

````python
class Node:
    def __init__(self):
        self.val = 0
        self.grad = [0, 0]
        self.lap = 0


class Mesh:
    def __init__(self, x_length, dx, y_length=None):
        self.x_length = x_length
        if y_length is None:
            self.y_length = x_length
        else:
            self.y_length = y_length

        self.dx = dx
        self.N_x = int(self.x_length / dx)
        self.N_y = int(self.y_length / dx)

        self.nodes: list[list[Node]] = []
        for i in range(self.N_x):
            row: list[Node] = []
            for j in range(self.N_y):
                row.append(Node())
            self.nodes.append(row)

    def assign_val(self, val_func):
        for i in range(self.N_x):
            for j in range(self.N_y):
                self.nodes[i][j].val = val_func(i * self.dx, j * self.dx)

    def calc_grad(self):
        for i in range(self.N_x):
            for j in range(self.N_y):
                # X direction gradient calculation:
                # Boundary Check:
                if i == 0:
                    self.nodes[i][j].grad[0] = (
                        self.nodes[i + 1][j].val - self.nodes[self.N_x - 1][j].val
                    ) / (2 * self.dx)
                elif i == self.N_x - 1:
                    self.nodes[i][j].grad[0] = (
                        self.nodes[0][j].val - self.nodes[i - 1][j].val
                    ) / (2 * self.dx)
                # Normal:
                else:
                    self.nodes[i][j].grad[0] = (
                        self.nodes[i + 1][j].val - self.nodes[i - 1][j].val
                    ) / (2 * self.dx)
                # Y direction gradient calculation:
                if j == 0:
                    self.nodes[i][j].grad[1] = (
                        self.nodes[i][j + 1].val - self.nodes[i][self.N_y - 1].val
                    ) / (2 * self.dx)
                elif j == self.N_y - 1:
                    self.nodes[i][j].grad[1] = (
                        self.nodes[i][0].val - self.nodes[i][j - 1].val
                    ) / (2 * self.dx)
                # Normal:
                else:
                    self.nodes[i][j].grad[1] = (
                        self.nodes[i][j + 1].val - self.nodes[i][j - 1].val
                    ) / (2 * self.dx)

    def calc_lap(self):
        for i in range(self.N_x):
            for j in range(self.N_y):

                # Boundary Check:
                if i == 0:
                    node_left = self.nodes[self.N_x - 1][j]
                else:
                    node_left = self.nodes[i - 1][j]
                if i == self.N_x - 1:
                    node_right = self.nodes[0][j]
                else:
                    node_right = self.nodes[i + 1][j]

                if j == 0:
                    node_down = self.nodes[i][self.N_y - 1]
                else:
                    node_down = self.nodes[i][j - 1]
                if j == self.N_y - 1:
                    node_up = self.nodes[i][0]
                else:
                    node_up = self.nodes[i][j + 1]
                    
                self.nodes[i][j].lap = (
                    node_up.val
                    + node_down.val
                    + node_left.val
                    + node_right.val
                    - 4 * self.nodes[i][j].val
                ) / (self.dx * self.dx)
````

So we added a new class, `Node`, which contains three properties:

- val: store the value;
- grad: a list to store the gradient in x and y direction;
- lap: store the laplacian value;

And, the class `Mesh` was modified that `nodes` not just 2D list filled with plain float number, but filled with `Node` object. To make it clearer, these properties are denoted type. Then `Mesh` have two new method now:

- cal_grad: calculate the gradient and store the value inside the node relatively.
- cal_lap: almost the same as above, but calculate laplacian.

The gradient and laplacian:

$$\nabla f = \left[\frac{\partial f}{\partial x} , \frac{\partial f}{\partial y} \right]^{\mathrm{T}}, $$
$$\nabla^2 f = \nabla \cdot \nabla f = \partial^2f/\partial x^2 + \partial^2f/\partial y^2 ;$$

And calculate them *at one point*
$$ (\nabla f)_{i,j} = \left[\frac{f_{i+1,j} - f_{i-1,j}}{2h},\frac{f_{i,j+1} - f_{i,j-1}}{2h}\right] , $$
$$ (\nabla^2 f)_{i,j} =  \frac{(f_{i+1,j}-f_{i,j})/h - (f_{i,j} - f_{i-1,j})/h}{h} + \frac{(f_{i,j+1}-f_{i,j})/h - (f_{i,j} - f_{i,j-1})/h}{h}$$
$$\\ = \frac{f_{i+1,j} - 2f_{i,j} + f_{i-1,j}}{h^2} + \frac{f_{i,j+1} - 2f_{i,j} + f_{i,j-1}}{h^2}$$
$$\\ = \frac{f_{i+1,j} + f_{i-1,j} + f_{i,j+1}  + f_{i,j-1} - 4f_{i,j} }{h^2}$$

Those who are familiar with Python must have noticed that there is actually no need for lower boundary check if we just adopt the periodical condition. Python will automatically consider list[-1] as the last one element. However, for the coherency and transferability  of the code, let's just use it.

Or let's see another function's laplacian: normal distribution's laplacian.

````python
x_axis_norm = np.linspace(0.1,mesh_2.x_length-0.1,mesh_2.N_x)
y_axis_norm = np.linspace(0.1,mesh_2.y_length-0.1,mesh_2.N_y)
X,Y = np.meshgrid(x_axis_norm, y_axis_norm)
## Here I use list comprehension to extract the property 'val' I need, and use one-liner if else statement.
## Due to the "periodic" boundary (of course here it should not be periodic boundary), the abnormal big value are filtratec.
Z_norm = np.array([[(np.nan if (o.lap>1000 or o.lap < -1000) else o.lap) for o in o1] for o1 in mesh_2.nodes])
Z_norm_val = np.array([[o.val for o in o1] for o1 in mesh_2.nodes])
fig = plt.figure(figsize=(18,8))

# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 2, 1, projection='3d')
p2 = ax.plot_surface(X, Y, Z_norm_val, rstride=1, cstride=1, cmap=coolwarm, linewidth=0, antialiased=False)
cb2 = fig.colorbar(p2, shrink=0.5)
ax = fig.add_subplot(1, 2, 2, projection='3d')
p1 = ax.plot_surface(X, Y, Z_norm, rstride=1, cstride=1, cmap=coolwarm, linewidth=0, antialiased=False)
cb1 = fig.colorbar(p1, shrink=0.5)
````

Hope this time it can be more clearer for you to visualize the result. When the function "changing its variation rate dramatically", its laplacian will reach maxima or minima. When the function "vary in constant rate", the laplacian will also vary in a somehow constant rate.
